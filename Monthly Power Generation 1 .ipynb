# Monthly Power Generation of Uttar Pradesh

### Overview: 
This dataset captures the total monthly power generation in Uttar Pradesh, India, across various power stations. It spans from 2008 to 2017, providing insights into power production trends over time. The data is sourced from the Ministry of Power, Government of India and is hosted by NDAP (National Data & Analytics Platform).

### Why this specific dataset?

1. Rich Time-Series Data: The dataset covers monthly power generation from 2008 to 2017, making it ideal for trend analysis.

2. Seasonal Variations: Power demand fluctuates due to weather conditions, industrial cycles, and policy changes, allowing us to study seasonality.

3. Trend & Forecasting: Helps in identifying long-term growth, periodic patterns, and future projections using statistical models.

4. Real-World Relevance: Supports energy sector analysis, resource planning, and decision-making, making it an excellent case study.

This dataset provides a comprehensive foundation for applying trend analysis, moving averages, and regression techniques.

## Moving Average Method

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file
df = pd.read_csv("Documents/MonthlyPowerGeneration.csv")
df.index = df.index + 1 #Change index to start from 1
# Display the dataset
df
```

```python
# Initialize moving average list with first 6 values as NaN
moving_avg = [None] * 6  

# Compute moving average
for i in range(7, len(df) - 5):  
    A1 = sum(df.iloc[i-7:i+5, 1]) / 12  
    A2 = sum(df.iloc[i-6:i+6, 1]) / 12  
    moving_avg.append((A1 + A2) / 2)  

# Fill last 6 entries as NaN
moving_avg += [None] * 6
```

```python
# Add the moving average column
df["Moving Average"] = moving_avg[:len(df)]
pd.set_option('display.max_rows', None)
df
```

```python
# Plot original data
plt.figure(figsize=(15, 6))
plt.plot(df["Month"], df.iloc[:, 1], label="Monthly Power Generation", color='blue', linestyle='dashed', marker='o')

# Plot moving average
plt.plot(df["Month"], df["Moving Average"], label="Trend", color='red', linewidth=2)

# Labels and title
plt.xticks(df["Month"][::2], rotation=90)
plt.xlabel("Time (Months)")
plt.ylabel("Power Generation")
plt.title("Time Series with 12-Month Moving Average")
plt.legend()

# Show the plot
plt.show()
```

## Exponential Smoothing Method

```python
# Define smoothing factor (alpha)
alpha = 0.2  

# First value of Tn is set to X1
Tn = [df.iloc[0, 1]]

# Compute Tn for rest of the data
for i in range(1, len(df)):
    Tn.append(alpha * df.iloc[i, 1] + (1 - alpha) * Tn[-1])

# Add Tn column to DataFrame
df["Tn"] = Tn

# Display the DataFrame
pd.set_option('display.max_rows', None)
df
```

```python
# Plot original data
plt.figure(figsize=(15, 6))
plt.plot(df["Month"], df.iloc[:, 1], label="Monthly Power Generation", color='blue', linestyle='dashed', marker='o')

# Plot moving average
plt.plot(df["Month"], df["Moving Average"], label="Trend", color='red', linewidth=2)

# Plot exponential smoothing
plt.plot(df["Month"], df["Tn"], label="Exponential Smoothing (=0.2)", color='green', linewidth=2)

# Labels and title
plt.xticks(df["Month"][::2], rotation=90)
plt.xlabel("Time (Months)")
plt.ylabel("Power Generation")
plt.title("Time Series with 12-Month Moving Average and Exponential Smoothing")
plt.legend()

plt.show()
```

## Quadratic Trend 

```python
import numpy as np

# Step 1: Define dependent variable (Y)
Y = np.array(df.iloc[:, 1]).reshape(-1, 1)

# Step 2: Define independent variables (X)
n = len(Y)
X1 = np.ones((n, 1)) 
X2 = np.arange(1, n + 1).reshape(-1, 1) 
X3 = X2**2

# Combine all into matrix X
X = np.hstack((X1, X2, X3)).astype(int)
```

```python
# Step 3: Compute beta coefficients using matrix formula  = (X'X)^(-1) X'Y
beta_hat = np.linalg.inv(X.T @ X) @ X.T @ Y 

# Step 4: Calculate predicted values and residuals
Y_hat = X @ beta_hat  # Predicted values
U_hat = Y - Y_hat  # Residuals

beta_hat
```

```python
# Step 5: Compute RSS & Variance

k = X.shape[1]  # Number of parameters
RSS = U_hat.T @ U_hat  # Residual sum of squares
sigma_sq = (RSS / (n - k))  # Residual variance
Var = sigma_sq * np.linalg.inv(X.T @ X)  # Variance-covariance matrix
```

```python
# Step 6: Compute test statistics

SE = np.sqrt(np.diag(Var)).reshape(-1, 1)
t_values = beta_hat / SE

t_values
```

```python
# Step 7: Compare with t-critical value

import scipy.stats as stats
alpha = 0.05  # Significance level
DoF = n-k  # Degrees of freedom
t_critical = stats.t.ppf(1 - alpha/2, DoF)
significance = np.abs(t_values) > t_critical
significance
```
